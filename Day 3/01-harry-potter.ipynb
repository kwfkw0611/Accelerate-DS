````python
# list of important words
people = ['Harry', 'Ron', 'Voldemort', 'Sirius', 'Vernon', 'Hermione']
words = ['Azkaban', 'Hogwarts', 'Gryffindor', 'Hufflepuff', 'Ravenclaw', 'Slytherin']
````

````python
# function to count the frequency of importants word in harry potter text file into a dictionary
def word_freqs(lst, file):
    """ function takes 2 arguments, a list of important words and splitted text file """
    freq = {}
    for i in file:
        if i in lst:
            if i in freq:
                freq[i] += 1
            else:
                freq[i] = 1        
    return freq
````

````python
# frequency of important characters
chars = word_freqs(people, s)
chars
````

````python
# frequency of important places 
houses = word_freqs(words, s)
houses
````

## Chapters

1.) Split the text into chapters.

```python
chapters = text.split("CHAPTER")

```

For each chapter count:

* The number of sentences 
* The number of words 
* How often each person is mentioned.

````python
# counts = {"Number of Sentences": text.count("."), "Number of Words": len(text.split())}
````

````python
# splitting text into list of chapters
chapters = text.split("CHAPTER")

# index 0 is title
chapters = chapters[1:]
````

````python
# for loop to count the number of chapters and number of sentences and words in each chapter into a dictionary of dictionary
chap_counts = {}
for n in range(len(chapters)):
    chap_counts[n+1] = {'Sentences': chapters[n].count("."), "words": len(chapters[n].split())}
````

````python
chap_counts
````

````python
# list of chapters splitted on whitespace
chp_clean = []
for i in chapters:
    x = i.split()
    chp_clean.append(x)
````

````python
# count frequency of important characters in each dictionary into a dictionary  of dictionary
chap_names = {}
for n in range(len(chapters)):
    chap_names[n+1] = word_freqs(people, chp_clean[n])
````

````python
chap_names
````

````python
# count frequency of important places in each chapter into a dictionary of dictionary
chap_houses = {}
for n in range(len(chapters)):
    chap_houses[n+1] = word_freqs(houses, chp_clean[n])
````

````python
chap_houses
````

````python
# convert dict of dict to dataframe for important names
df_chp_names = pd.DataFrame.from_dict(chap_names, orient='index')
````

````python
# convert dict of dict to dataframe for important places
df_chp_houses = pd.DataFrame.from_dict(chap_houses, orient='index')
````

````python
# list of dataframes for concat
frames = [df_chp_names, df_chp_houses]
````

````python
# concat dataframes column wise
df_chapters = pd.concat(frames, axis =1)
````

````python
df_chapters
````

````python
# fill NaN values with 0
df_chapters = df_chapters.fillna(value = 0)
````

````python
df_chapters.head()
````

````python
# naming index 
df_chapters.index.name = 'Chapters'
````

````python
df_chapters.head()
````

````python
# line plot 
df_chapters.plot()
plt.title("Popular Words in Harry Potter per Chapter")
````
